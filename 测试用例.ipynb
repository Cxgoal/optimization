```python
import numpy as np
import time
import pandas as pd
from optimizer.trustRegionMethod import TrustRegion
from testFunction import ExtendedPowellSingular, Rosenbrock, BiggsEXP6, PowellBadlyScaled
```

```python
def test(test_function, method, start_point):
    s_time = time.time()
    x = method.compute(test_function.f, test_function.g, test_function.gg, start_point)
    e_time = time.time()
    return x, test_function.f(x), method.iter_num, e_time-s_time
```

```python
iter_a = []
x_a = []
f_a = []
t_a = []
method_a = ["hebden"]

for m in [8, 9, 10, 11, 12]:
    f = BiggsEXP6(m)
    fx0 = np.array([1, 2, 1, 1, 1, 1]).reshape(6, 1)
    for subq in method_a:
        method = TrustRegion(subq_method=subq, delta=0.1, max_iter=2000, max_error=1e-8)
        res = test(f, method, fx0)
        x_a.append(res[0])
        f_a.append(res[1])
        iter_a.append(res[2])
        t_a.append(res[3])
t = pd.DataFrame()
t['x'] = x_a
t['fx']=f_a
t['iter'] = iter_a
t['t'] = t_a
```
